# Modeling

```python
features = ['lot_area', '1st_flr_sf', '2nd_flr_sf', 'full_bath', 
            'half_bath', 'pool_area', 'total_bsmt_sf', 'garage_area']

X = train[features]
y = train['saleprice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_train, y_train), lr.score(X_test, y_test)
```

```python
preds = lr.predict(test[features])
```

```python
test['saleprice'] = preds
```

```python
test_final = test[['id', 'saleprice']]
```

```python
test_final.to_csv('datasets/test_mlr.csv', index = False)
```
